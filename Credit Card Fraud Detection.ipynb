```python
# importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# getting the rows & column of data
df.shape
```

```python
# getting the 10% of data to reduce computational difficulty
df = df.sample(frac=0.1, random_state=0)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
# getting the histogram of df
df.hist(figsize=(20,20))
plt.show()
```

```python
# getting the features of df
df.columns
```

```python
# In Class 0 - genuine transaction; 1 - fraud transaction
fraud_df = df[df['Class']==1]
genuine_df = df[df['Class']==0]
```

```python
# getting the fraud transaction percentage
fraud_fraction = len(fraud_df) / float(len(genuine_df))
print('Percentage of fraud tansaction is {}%'.format(round(fraud_fraction*100, 2)))
```

```python
# Correlation
correlation = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation)
plt.show()
```

```python
# Separating the dependent and independent features
independent_feature = df.columns.to_list()[:-1]
dependent_feature = ['Class']
X = df[independent_feature]
y = df[dependent_feature]
```

```python
# importing the algorithm

from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

Classifier = {
    'IsolationForest': IsolationForest(max_samples=len(X), contamination=fraud_fraction, random_state=0),
    'LocalOutlierFactor': LocalOutlierFactor(n_neighbors=20, contamination=fraud_fraction)
}
```

```python
Classifier
```

```python
# fitting the model
for class_name, classi in Classifier.items():
    if class_name == 'LocalOutlierFactor':
        y_pred = classi.fit_predict(X)
    else:
        y_pred =classi.fit_predict(X)
    
    # Reconfiguring the label as 0 - valid and 1 - fraud
    y_pred[y_pred==1] = 0
    y_pred[y_pred==-1] = 1
    
    n_error = (y_pred != y.to_numpy).sum()
    
    # Classification report
    
    print('{}: {}'.format(class_name, n_error))
    print(accuracy_score(y, y_pred))
    print(classification_report(y,y_pred))
```
